```python
corr_df = df_tracks.select_dtypes(include=['number']).corr(method="pearson")

plt.figure(figsize=(14,6))
heatmap = sns.heatmap(
    corr_df,
    annot=True,
    fmt=".1g",
    vmin=-1,
    vmax=1,
    center=0,
    cmap="inferno",
    linewidths=1,
    linecolor="black"
)

heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)

```

```python
# Create a clean 'dates' column from index (handle mixed date formats)
df_tracks["dates"] = pd.to_datetime(
    df_tracks.index.get_level_values("release_date"),
    errors='coerce',          # skip invalid dates instead of crashing
    format='mixed'            # allows year-only or year-month-day
)

# Extract year
df_tracks["year"] = df_tracks["dates"].dt.year

# Group by year to avoid duplicates
yearly_duration = df_tracks.groupby("year")["duration"].mean().reset_index()

# Plot
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x="year", y="duration", data=yearly_duration, ax=ax, errorbar=None)
ax.set_title("Year vs Average Song Duration")
plt.xticks(rotation=60)
plt.show()
```

```python
# Clean date column from index (handles year-only and full dates)
df_tracks["dates"] = pd.to_datetime(
    df_tracks.index.get_level_values("release_date"),
    errors='coerce',
    format='mixed'
)

# Extract year
df_tracks["year"] = df_tracks["dates"].dt.year

# Group by year for average duration
yearly_duration = df_tracks.groupby("year")["duration"].mean().reset_index()

# Plot line graph
fig_dims = (18, 7)
plt.figure(figsize=fig_dims)
sns.lineplot(data=yearly_duration, x="year", y="duration", marker="o", linewidth=2)
plt.title("Year vs Average Song Duration", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Average Duration (ms)")
plt.xticks(rotation=60)
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```
